```python
import pandas as pd
path = '/content/Lakers 2021-2022 Data Sheet(AutoRecovered).csv'
data = pd.read_csv(path)
df = pd.DataFrame(data, columns = ['Lineups', 'Plus minus', 'FGA', 'FGM', 'FG %', 'True Field Goal  %', '3PM', '3PA', '3P%', 'Offensive Rebounds', 'Assists', 'TO', 'As/TO ratio', 'Possessions', 'Points Scored', 'OEF', 'FTA', 'FTM', 'Minutes'])

print("Welcome to my BIOL 222 Project. In this program, you can sort the data I collected in however you would like to in order to see what lineup works best for specific players based on specific criteria")
print("The first part of my code is picking the players you would like to sort by")
print("In order to do so, you must type the abbreviations of player's names according to the legend below:")
print("Avery Bradley = (AB)")
print("Anthony Davis = (AD)")
print("Austin Reaves = (AR)")
print("Carmelo Anthony = C")
print("Deandre Jordan = D")
print("Wayne Ellington = E")
print("Dwight Howard = H")
print("Jay Huff = J")
print("Kent Bazemore = K")
print("Lebron James = L")
print("Malik Monk = M")
print("Rajon Rondo = R")
print("Sekou Doumbouya = (SD)")
print("Talen Horton Tucker = T")
print("Russell Westbrook = W")


numberOfPlayers = int(input("How many players would you like to sort by?"))
i = 0
listOfPlayers = []
while i < numberOfPlayers:
  P = input("What player(s) would you like to sort by?")
  listOfPlayers.append(P)
  i += 1

listOfPlayers.append("Average")
Sortify = input("What statistic would you like to sort by?")


i = 0
while i < len(listOfPlayers):
  df = df[df["Lineups"].str.contains(listOfPlayers[i], na=False)]
  i += 1

sortedData = df.sort_values(by=[Sortify], ascending=False)
print(sortedData)


```
